# 1 - pegar tudo do html

```python
#URL to scrape
#url = 'https://www.wikiwand.com/en/Kylie_Minogue_singles_discography'
```

```python
url = 'https://en.wikipedia.org/wiki/Kylie_Minogue_singles_discography'
```

```python
df = pd.read_html(url)
```

```python
df_80 = df[1]
```

```python
df_80["Decade"] = "80"
```

```python
df_80.head()
```

```python
#identificar todas as colunas
df_80.columns
```

```python
#limpar nomes das colunas
df_80.columns = ['Title',
            'Year',
            'Peak_AUS',
            'Peak_AUT',
            'Peak_FRA',
            'Peak_GER',
            'Peak_IRE',
            'Peak_NZ',
            'Peak_SWI',
            'Peak_UK',
            'Peak_US',
            'Peak_USDance',
            'Sales',
            'Certifications',
            'Album',
            'Unnamed',
            'Decade']
```

```python
df_80.head()
```

```python
#limpar colunas (drop Peak_USDance, Sales, Certifications, Unnamed)
df_80 = df_80.drop(['Peak_USDance', 'Sales', 'Certifications', 'Unnamed'], axis = 1)
#dropou inplace
```
