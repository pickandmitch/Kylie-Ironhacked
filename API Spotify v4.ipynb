# conectar com Spotify

```python
import pandas as pd
import os

from spotipy import util, oauth2, Spotify

from dotenv import load_dotenv

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
spotify_client_id = 'deletado'
spotify_client_secret = 'deletado'
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=spotify_client_id,client_secret=spotify_client_secret))
```

# pegar o ID da artista

```python
sp.artist('4RVnAU35WRWra6OZ3CbbMA')
```

# pegar os albuns e seus IDs

```python
#get kylies albums
discos = sp.artist_albums('4RVnAU35WRWra6OZ3CbbMA', 
                            album_type = ('album' or 'single'), 
                              country = 'GB')
```

```python
disco_id = []
disco_name = []
disco_release = []
#disco_markets = []

for x in discos['items']:
    disco_id.append(x['id'])
    disco_name.append(x['name'])
    disco_release.append(x['release_date'])
    #disco_markets.append(x['available_markets'])
```

```python
df_albums = pd.DataFrame({'disco_id': disco_id,
        'disco_name': disco_name,
        'disco_release': disco_release})
```

```python
print(disco_id)
```

# pegar as musicas a partir dos albuns

```python
track_id = []
track_name = []
track_disco_id = []

for x in disco_id:
    results_discos = sp.album_tracks(x)['items']
    
    for y in results_discos:
        track_id.append(y['id'])
        track_name.append(y['name'])
        track_disco_id.append(x)
```

```python
len(track_disco_id)
```

```python
len(track_name)
```

```python
print(track_name)
```

# pegar os audio features a partir dos IDs das musicas

```python
track_danceability = []
track_energy = []
track_tempo = []
track_loudness = []

for id_musica in track_id:
    print(id_musica)   
    track_DJ = sp.audio_features(id_musica)
    track_danceability.append(track_DJ[0]['danceability'])
    track_energy.append(track_DJ[0]['energy'])
    track_tempo.append(track_DJ[0]['tempo'])
    track_loudness.append(track_DJ[0]['loudness'])
    
```

```python
df_tracks = pd.DataFrame({'track_id': track_id,
        'track_name': track_name,
        'track_danceability': track_danceability,
        'track_energy': track_energy,
        'track_temp': track_tempo, 
        'track_loudness': track_loudness,
        'track_disco_id': track_disco_id})
```

```python
df_tracks.head()
```

```python
df_merged = df_tracks.merge(df_albums, left_on = 'track_disco_id', right_on = 'disco_id')
```

```python
df_merged.to_csv('df_merged.csv', index = False)
```
